```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        
        sumNode = None
        sumNodeTail = None
        carry = 0
        newSum = 0
        i = l1
        j = l2
        i_val = 0
        j_val = 0
        while (i != None or j != None) :
            # extract val
            if i == None:
                i_val = 0
            else:
                i_val = i.val
            
            if j == None:
                j_val = 0
            else:
                j_val = j.val
            
            # now calculate the sum 
            newSum = i_val + j_val + carry 
            newDigit = newSum % 10
            # update carry
            if newSum >= 10:
                carry = 1
            else:
                carry = 0
                
            # update node      tail 
            if sumNode == None:
                sumNode = ListNode(newDigit)
                sumNodeTail = sumNode
            else:
                newDigNode = ListNode(newDigit)
                sumNodeTail.next = newDigNode
                sumNodeTail = sumNodeTail.next
            
            # move i and j forward 
            if (i != None):
                i = i.next
            
            if (j != None):
                j = j.next
            
            # corner case need to add one more node 
            if i == None and j == None and carry == 1:
                lastDigNode = ListNode(1)
                sumNodeTail.next = lastDigNode
            
            # update i and v value
        return sumNode
            
            
            
            
            
            
            
            
            
            
            
            
            
            
```
