Given two binary trees and imagine that when you put one of them to cover the other, some nodes of the two trees are overlapped while the others are not.

You need to merge them into a new binary tree. The merge rule is that if two nodes overlap, then sum node values up as the new value of the merged node. Otherwise, the NOT null node will be used as the node of new tree.

```python
# My BFS solution is too slow but it works
class Solution:
    def mergeTrees(self, t1, t2):
        """
        :type t1: TreeNode
        :type t2: TreeNode
        :rtype: TreeNode
        """
    
        if not t1 :
            return t2
        if not t2:
            return t1
        
        q1 = [t1]
        q2 = [t2]
        combineQ = []
        # traverse the tree to merge 
        while (len(q1) > 0) and (len(q2) > 0):
            q1Node = None
            q2Node = None
            # this should be exact same node on both tree
            totalVal = None
            if len(q1):
                q1Node = q1.pop(0)
            if len(q2):    
                q2Node = q2.pop(0)
            
            # check to see if what we should push
            if q1Node:
                if not totalVal:
                    totalVal = 0
                totalVal += q1Node.val 
                # push the child 
                q1.append(q1Node.left)
                q1.append(q1Node.right)
                # to match with the level of q1
                if not q2Node:
                    q2.append(None)
                    q2.append(None)
            if q2Node:
                if not totalVal:
                    totalVal = 0
                totalVal += q2Node.val 
                # push the child 
                q2.append(q2Node.left)
                q2.append(q2Node.right)
                # to match with the level of q1
                if not q1Node:
                    q1.append(None)
                    q1.append(None)
                
            # now merge two node if both are
            if totalVal != None :
                newNode = TreeNode(totalVal)
                combineQ.append(newNode)
            else:
                combineQ.append(None)
        
        # now build the tree
        root = combineQ[0]
        #print(combineQ)
        bfsQ = [combineQ.pop(0)]
        while (len(combineQ) > 0) and (len(bfsQ) > 0):
            rootNode = bfsQ.pop(0)
            if rootNode:
                rootNode.left = combineQ.pop(0)
                rootNode.right = combineQ.pop(0)
                bfsQ.append(rootNode.left)
                bfsQ.append(rootNode.right)
        
        return root
    
```

```python
# My BFS solution is too slow but it works
class Solution:
    def mergeTrees(self, t1, t2):
        """
        :type t1: TreeNode
        :type t2: TreeNode
        :rtype: TreeNode
        """
        # nothing to merge 
        if not t1 :
            return t2
        if not t2:
            return t1
        
        # the key idea here is pick one tree to merge the other 
        # and we can either go level by level or by depth
        # here we will use BFS as well as the node of one node match the other 
        # also we need to push and pop the similar position nodes on both tree
        
        queue = [(t1,t2)] # each tuple will represent nodes at the same position each trees
        # traverse while queue is not empty
        # pick t1 to be merged
        while len(queue):
            t1Node, t2Node = queue.pop(0)
            t1Node.val = t1Node.val + t2Node.val
            #check left tree
            # case 1: take whatever from t2
            if not t1Node.left:
                t1Node.left = t2Node.left
            # case 2: push the same node to the tree
            elif (t1Node.left) and (t2Node.left):
                queue.append((t1Node.left, t2Node.left))
            
            # check right tree
            # case 1: t1Node is null then take whatever from t2Node
            if not t1Node.right:
                t1Node.right = t2Node.right
            # push the same node to the queue     
            elif (t1Node.right) and (t2Node.right):
                queue.append((t1Node.right, t2Node.right))
        
        return t1
           
            
```

```python
# TREE HELPERS 
def stringToTreeNode(input):
    input = input.strip()
    input = input[1:-1]
    if not input:
        return None

    inputValues = [s.strip() for s in input.split(',')]
    root = TreeNode(int(inputValues[0]))
    nodeQueue = [root]
    front = 0
    index = 1
    while index < len(inputValues):
        node = nodeQueue[front]
        front = front + 1

        item = inputValues[index]
        index = index + 1
        if item != "null":
            leftNumber = int(item)
            node.left = TreeNode(leftNumber)
            nodeQueue.append(node.left)

        if index >= len(inputValues):
            break

        item = inputValues[index]
        index = index + 1
        if item != "null":
            rightNumber = int(item)
            node.right = TreeNode(rightNumber)
            nodeQueue.append(node.right)
    return root

def treeNodeToString(root):
    if not root:
        return "[]"
    output = ""
    queue = [root]
    current = 0
    while current != len(queue):
        node = queue[current]
        current = current + 1

        if not node:
            output += "null, "
            continue

        output += str(node.val) + ", "
        queue.append(node.left)
        queue.append(node.right)
    return "[" + output[:-2] + "]"
```
