## MEREGE SORT 

```python
# merge two array S1, S2 into S
# assume S1 is left and S2 is right side
def merge (S1, S2, S):
    if len(S) < (len(S1) + len(S2)):
        raise Exception()
    n = len(S)
    i = 0
    j = 0
    while (i + j) < n:
        if (j == len(S2)) or (i < len(S1) and (S1[i] < S2[j])):
            S[i+j] = S1[i]
            i += 1
        else:
            S[i+j] = S2[j]
            j += 1

#mereg sort call 
def merge_sort (S):
    n = len(S)
    if n <= 1:
        return S
    mid = n // 2
    S1 = S[0:mid]
    S2 = S[mid:n]
    # divide the work
    merge_sort(S1)
    merge_sort(S2)
    # merge both sides to original 
    merge(S1, S2, S)
```
